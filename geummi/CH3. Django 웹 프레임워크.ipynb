* 개발 단위   
1. 프로젝트 : 웹 사이트에 대한 전체 프로그램   
2. 애플리케이션 : 모듈화 된 단위 프로그램   
**애플리케이션 프로그램들이 모여서 프로젝트가 개발됨**

* MVT 패턴에 따른 애플리케이션 개발
: MVC(Model-View-Controller)와 같은 개념으로, 데이터(Model), 사용자 인터페이스(VIew), 데이터를 처리하는 로직(Controller)을 구분하여 한 요소가 다른 요소에 영향을 주지 않도록 설계함   
> Django에서는 View를 Template, Controller를 View라고 표현   

1. 클라이언트로부터 요청을 받으면 URLconf를 이용하여 URL을 분석함   
2. URL 분석 결과를 통해 해당 URL에 대한 처리를 담당할 뷰를 결정함   
3. 뷰는 자신의 로직을 실행하면서, 만일 데이터베이스 처리가 필요하면 **모델**을 이용해 처리하고, 그 결과를 반환받음   
4. 뷰는 자신의 로직 처리가 끝나면, **템플릿**을 사용하여 클라이언트에 전송할 HTML 파일을 생성함   
5. 뷰는 최종 결과로 HTML 파일을 클라이언트에게 보내 응답함

## 2. URLconf (URL 정의)
:클라이언트의 요청에서 URL 분석에 사용되는 URL 패턴 정의  
> **path('호출하려는 url', 처리함수),**   
> <<int:year>> 안과 입력한 url의 타입이 일치하지 않으면 에러   

> * str : 디폴트 타입, 슬래시를 제외한 문자열   
> * int : 0 또는 양의 정수   
> * slug : ASCII,숫자, 하이픈 형식의 문자열   
> * uuid : UUID 타입의 문자열
> * path : 슬래시를 포함한 모든 문자열과 매치
