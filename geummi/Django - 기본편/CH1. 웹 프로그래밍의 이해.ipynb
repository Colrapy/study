### 2. HTTP 처리 방식

1. GET   
: 가장 많이 이용되는 메소드   
  지정한 URL의 정보를 가져오는 역할   
  웹 브라우저를 이용하여 서버로부터 정보를 가져오려고 할 때 이용   
 > https://www.naver.com?이름 = 값
 
2. POST   
: 리소스의 생성   
  리소스의 생성은 PUT도 할 수 있지만, 차이점은 URL의 프레임 결정권이 클라이언트에게 있을 경우는 PUT을 이용하며 그렇지 않은 경우에는 POST릉 이용   
  보통 GET과 동시에 사용할 수 있는 역할인 경우 GET을 많이 이용   
 > https://www.naver.com/프토토콜/상태메세지   
   이름 = 값
   > GET과 달리 바디에 정보를 숨겨서 보안이 높음
   
3. PUT   
: 리소스의 변경   
  URL 프레임 결정권이 클라이언트에 있을 경우, 리소스의 생성


4. DELETE   
: 리소스의 삭제   
  바디를 반환하지 않고 응답

### 4. URL의 설계

* URL   
: 웹 서버에 존재하는 애플리케이션에 대한 API   

1. RPC (Remote Procedure Call)   
: 클라이언트가 네트워크 상에서 원격에 있는 서버가 제공하는 API 함수를 호출   
  URL 설계와 API 설계를 동일하게 고려   
  URL 경로의 대부분을 동사로 구성   
> http://blog.example.com/search?q=test&debug=true   
    > 내용을 숨기지 않음   
    
2. REST (Representatinal State Transfer)   
: 웹 클라이언트에서 URL을 전송하는 것이 웹 서버에 있는 리소스 상태에 대한 데이터를 주고받는 것   
  = 간단한 구조의 URL
> http://blog.example.com/search/test   
    > 내용을 간단한 단어로 대체함
