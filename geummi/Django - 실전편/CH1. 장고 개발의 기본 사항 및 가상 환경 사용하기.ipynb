# CH1. 장고 개발의 기본 사항 및 가상 환경 사용하기

## 장고의 특징

1. MVT 개발방식   
: Model - View - Template 로 나누어 개발을 진행하여 모듈 간에 독립성을 유지 가능하고 느슨한 결합 설계의 원칙에 부합함

> 개발 순서   
> 1. 프로젝트 뼈대 만들기 : 프로젝트 및 앱 개발에 필요한 디렉터리와 파일 생성   
> 2. 모델 코딩하기 : 테이블 관련 사항을 개발 (models.py, admin.py)   
> 3. URLconf 코딩하기 : URL 및 뷰 매핑 관계를 정의 (urls.py)   
> 4. 뷰 코딩하기 : 애플리케이션 로직 개발 (views.py)   
> 5. 템플릿 코딩하기 : 화면 UI 개발 (templates/*.html)

2. settings.py의 주요사항   
- 데이터베이스 설정   
- 애플리케이션 등록 : 프로젝트에 사용되는 모든 애플리케이션을 등록   
- 템플릿 항목 설정 : TEMPLATES 항목으로 지정   
- 정적 파일 항목 설정 : STATIC_URL 등의 항목 지정   
- 타임존 지정 : 기존 시간을 한국 시간으로 변경

3. models.py의 주요사항   
: ORM 기법을 사용하여, 테이블을 클래스로 매핑해서 테이블에 대한 CRUD 기능을 클래스 객체에 대해 수행하면, 장고가 내부적으로 SQL 처리를 하여 데이터베이스에 반영해줌
- django.db.models.Model 클래스를 상속받아 테이블을 생성함   
- 필드 클래스로 각 클래스 변수의 타입을 지정함   
- 변경사항 발생 시 makemigrations 및 migrate 실행함   

4. URLconf의 주요사항   
: 프로젝트 URL과 앱 URL을 나누어 2계층으로 코딩하여 변경이 쉽고 확장이 용이함   
- URL 패턴별 이름 지정이 가능함   
- 패턴 그룹별 이름공간 지정이 가능함   

5. views.py의 주요사항   
: 가독성과 유지보수 편리성, 재활용 등을 고려하며 코딩해야함   
- 함수형 뷰와 클래스형 뷰가 존재함   

6. templates의 주요사항   
: 웹 화면 별로 템플릿 파일이 하나씩 필요하고, 프로젝트 템플릿 디렉터리를 만들어 하나에 저장함
- base.html : 전체 프로젝트의 룩앤필에 관련한 파일을 모아둠   
- 각 앱에서 사용하는 템플릿 파일들은 앱 템플릿 디렉터리에 위치시킴

> 템플릿 디렉터리 예시   
> - 프로젝트 루트 디렉터리 : /home/hgm/study/ch1   
> - 프로젝트 디렉토리 : /home/hgm/study/ch1/mysite   
> - 프로젝트 템플릿 디렉터리 : /home/hgm/study/ch1/templates   
> - 앱 템플릿 디렉터리 : /home/hgm/study/ch1/bookmark/templates

7. Admin 사이트   
: 테이블의 내용을 열람하고 수정하는 기능을 제공

8. runserver   
: 개발 과정에 현재의 웹 프로그램을 실행해볼 수 있도록 테스트용 웹 서버를 제공함
- 상용화된 Apache, Nginx 등의 사용 웹 서버 존재


## 가상환경 - VENV

1. 가상환경 만들기

```
# 파이썬 버전 확인
python3 --version

# 가상환경 모듈 생성
python3 -m venv vDjBook

# 가상환경 진입
source C:\study\Django_02\VENV\vDjBook\bin\activate

# 가상환경 탈출
deactivate
```

2. 가상환경에 패키지 설치하기

```
# 가상환경 진입
source C:\study\Django_02\VENV\vDjBook\bin\activate

# 태그 달기 패키지
pip3 install django-taggit
pip3 install django-taggit-templatetags2

# 폼 장식 패키지
pip3 install django-widget-tweaks

# 이미지 처리 패키지
pip3 install Pillow

# 타임존 관리 패키지
pip3 install pytz
```

3. 가상환경 패키지 툴 업그레이드
```
# 가상환경 진입
source C:\study\Django_02\VENV\vDjBook\bin\activate
# 업그레이드
pip3 install -U pip setuptools wheel
```

4. InsecurePlatformWarning 해결하기   
: 패키지 설치 과정에서 발생하는 경우, 해결 방법   
```
pip3 install pyopenssl ndg-httpsclient pyasn1
```
